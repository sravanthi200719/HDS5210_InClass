### 41.2 Plotting a KDE instead

Instead of a historgam, let's plot overlapping [KDEs](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.kde.html).

```python

graph = data[['IP Cash Price','IP Charge','IP Expected Reimbursement']].plot.kde()
graph.set_xlim(0, 40000)
```

### Question:

If I told you what these values represent, what conclusions could you draw about the pattern?
* IP Cash Price is the amount you would pay on your own without using insurance
* IP Charge is the maximum amount for which the hospital says it can submit a bill to an insurance company
* IP Expected Reimbursement is the amount that the hospital believes it is contracted to receive in payment from an insurance company

### YOUR RESPONSE:
1. IP Cash Price — Law of Diminishing Margins (green): the earliest curving and has least spreading, saying the best value is to pay cash on everything Then it more or less plummets to zero, which tells you that very few cash payers need ultra-high dollar procedures. Perhaps this is a pricing strategy aimed at self-pay patients, or who have high deductible plans.

2. IP Charge: This curve has the most coverage and is the flag bunch rightward, so indicating that this group are normally sent to Bit insurance bills the maximum amount. The long tail demonstrates that some procedures can have a very high max charge. This might be a play to establish an upper limit for negotiations with insurance companies, or a way of subsidizing the costliest cases.

3. IP Expected Reimbursement: This curve falls between the other two, maxing out higher than the cash before but lower than Connect. Implying that insurance companies, while able to negotiate rates lower than the maximum charges, generally pay more for most procedures compared with what a self-pay patient would be charged. ### 41.3 Creating a new set of series

In order to understand how this data varies across different payers, we have to create new data series that are based on the payer information.  In the data, we have a column called **Payer** that holds this information.

We can use the Pandas [groupby()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html) function to create groupings based on Payer and then plot kde based on those groupings.  When [plotting a Pandas groupby result](https://pandas.pydata.org/pandas-docs/version/0.16.2/generated/pandas.core.groupby.DataFrameGroupBy.plot.html), the options are different than plotting a dataframe.

```python
data['Payer'].value_counts()
```

```python
data.groupby(['Payer'])['IP Expected Reimbursement'].count()
```

```python
data.groupby(['Payer'])['IP Expected Reimbursement'].plot.kde(
    xlim=(0,40000),
    legend=True,
    figsize=(17,8.5)
)
```

### 41.4 Seeing if there's any correlation

So, we know that there are cash prices and insurance paid prices.  How strongly correlated are those amounts?

We can use a scatterplot with one value on the x axis and the other on the y axis to do a simple visualization of this.

```python
data.plot.scatter(x='IP Charge', y='IP Expected Reimbursement', color='DarkRed', label='Reimbursement')
```

```python
groups = data.groupby(['Payer'])
plt.figure(figsize=(17,11))

for i, (payer, prices) in enumerate(groups):
    ax = plt.subplot(3,3, i+1)
    prices.plot.scatter(x='IP Cash Price',y='IP Expected Reimbursement',ax=ax)
    ax.set_title(payer)
    ax.set_xlim(0,20000)
    ax.set_ylim(0,80000)

plt.tight_layout()
plt.show()
```

### YOUR RESPONSE:

Across all payers, the scatter plots show a usually positive association between IP Cash Price and IP Expected Reimbursement, although there are notable exceptions. Stronger, more linear correlations and larger reimbursement ranges—which frequently surpass cash prices—are displayed by private insurers such as Cigna and Blue Cross Blue Shield. Medicare and Medicaid, two government payers, on the other hand, show less consistent trends and pay out lower amounts, usually at or below cash prices. There are notable exceptions, especially for procedures with large reimbursements but modest cash prices.
